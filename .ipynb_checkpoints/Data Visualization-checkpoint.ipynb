Kernel Density Plots

```python
# Rugplot (marginal distribution) and histogram
ds = randn(100)
sns.rugplot(ds) 
plt.hist(ds, alpha = 0.5)
plt.show()
```

```python
import numpy as np
# Generate 50 - 2 = 48 numbers between min and max value in dataset
x_axis = np.linspace(ds.min()-1, ds.max()+1, 50)
```

```python
# Gaussian bandwidth - Rule of Thumb
bandwidth = 1.06*ds.std()*(len(ds)**(-1/5))
```

```python
from scipy.stats import norm
kernels = []
# Iterate over all points in the dataset
for point in ds:
    # For each point, generate a gaussian kernel spread over specified bandwidth taking x values from x_axis
    # i.e. N(point, bandwidth) @ x = (values in x_axis)
    kernel = norm(point, bandwidth).pdf(x_axis)
    
    kernel = kernel*0.5
    
    # Save each kernel
    kernels.append(kernel)
    
    plt.plot(x_axis, kernel, alpha = 0.5, color = "red")
```

```python
# Sum over entries at ith position
kde = np.sum(kernels, axis = 0)
```

```python
# Establishing kernel density plot
plt.plot(x_axis, kde, color = "green")
sns.rugplot(ds, color = "yellow")
plt.hist(ds, alpha = 0.5)
for kernel in kernels:
    plt.plot(x_axis, kernel, alpha = 0.5, color = "red")
```

```python
# Plotting KDE, Histogram and Rug-Plot using built-in function
sns.distplot(
    ds, 
    rug = True,
    rug_kws = { 
        "color" : "yellow" 
    }, 
    kde_kws = { 
        "color": "green" 
    },
    hist_kws = { 
        "color" : "blue" 
    },
)
```

```python
# Plotting Kernel Density Plot only
sns.kdeplot(ds, color = "green")
```

```python
# Bivariate Kernel Density Plots
import pandas as pd
mean = [0, 0]
covariance = [[3, 0],[0, 27]]
ds = np.random.multivariate_normal(mean, covariance, 500)
df = pd.DataFrame(ds, columns = ["column1", "column2"])
```

```python
sns.kdeplot(df, shade = True)
```

```python
sns.jointplot("column1", "column2", df, kind = "kde")
```

Boxplots

```python
sns.boxplot(ds, orient = "h")
```

Regression Plots

```python
df = sns.load_dataset("diamonds").sample(frac = 1).head(500)
sns.regplot("price", "carat", df, fit_reg = True)
```

```python
# Alternatively (Built on regplot and has more features)
sns.lmplot(
    "price", "carat", df, 
    scatter_kws = { 
#         "color" : "gray" 
    }, 
    line_kws = {
#         "color" : "blue"
    },
#   order = 2,
    fit_reg = False,
    hue = "cut",
    markers = ["^", "v", "*", ".", "s"],
    legend_out = False,
)
plt.legend(title = "Cut", loc = "upper left", edgecolor = "black")
```

Heatmaps

```python
df = sns.load_dataset("flights")
df2 = df.pivot(index = "year", columns = "month", values = "passengers")
```

```python
df2
```

```python
# %matplotlib notebook
sns.heatmap(
    df2, 
    annot = True, 
    annot_kws = {
        "size": 7
    }, 
    cmap = "YlGnBu", 
    fmt = "g",
)
```

```python
# Studying about a particular cell by assigning it the cenrer-most color from range
sns.heatmap(
    df2, 
    annot = True, 
    annot_kws = {
        "size": 7
    },
    cmap = "YlGnBu", 
    fmt = "g", 
    center = df2.loc[1955, "January"],
)
```

```python
# Cluster Map
sns.clustermap(
    df2, 
    annot = True, 
    annot_kws = {
        "size": 7
    },
    cmap = "YlGnBu", 
    fmt = "g",
#   standard_scale = 0 # Standardizing about row. Similarly 1 for column,
#   col_cluster = False, 
#   row_cluster = False,
)
```
